```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
import requests
import json
```

```python
def fetch_cpu_metrics(service, host, start, end, api_token):
    url = f"{host}/api/ds/query"
    expr = "sum(rate(container_cpu_usage_seconds_total{container=\""+service+"\"}[10m])) by (name) *100"
    body = {
        "queries": [
            {
                "editorMode": "code",
                "expr": expr,
                "legendFormat": "__auto",
                "range": True,
                "instant": False,
                "exemplar": False,
                "utcOffsetSec": 19800,
                "interval": "",
                "datasourceId": 1,
                "intervalMs": 10000,
                "maxDataPoints": 1791
            }
        ],
        "from": start,
        "to": end
    }
    response = requests.post(url, headers={"Content-Type": "application/json", "Authorization": f"Bearer {api_token}"}, data=json.dumps(body))
    json_response = json.loads(response.text)

    data = json_response["results"]["A"]["frames"][0]["data"]["values"]
    return data

def fetch_memory_metrics(service, host, start, end, api_token):
    url = f"{host}/api/ds/query"
    expr = "sum(rate(container_memory_usage_bytes{container=\""+service+"\"}[10m])) by (name) *100"
    body = {
        "queries": [
            {
                "editorMode": "code",
                "expr": expr,
                "legendFormat": "__auto",
                "range": True,
                "instant": False,
                "exemplar": False,
                "utcOffsetSec": 19800,
                "interval": "",
                "datasourceId": 1,
                "intervalMs": 10000,
                "maxDataPoints": 1791
            }
        ],
        "from": start,
        "to": end
    }
    response = requests.post(url, headers={"Content-Type": "application/json", "Authorization": f"Bearer {api_token}"}, data=json.dumps(body))
    json_response = json.loads(response.text)

    data = json_response["results"]["A"]["frames"][0]["data"]["values"]
    return data
```
